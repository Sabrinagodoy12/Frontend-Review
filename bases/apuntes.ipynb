# Operadores
- **Aritméticos:**

suma: + 

resta: - 

multiplicacion: * 

division:  / 

modulo: %

incremento: ++

decremento: --

- **Relacionales:**

Igualdad: ==

Igualdad estricta: ===

Desigualdad: !=

Desigualdad estricta: !==

Mayor y menor: ><

mayor o igual y menor o igual: >= <=

- **Condicionales o lógicos**

AND: && --> true si ambas son true

OR: || --> true solo si una es true

NOT: ! --> cambia el valor de verdad
